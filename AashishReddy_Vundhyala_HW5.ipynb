```python
#1.

%matplotlib notebook
%matplotlib notebook  
import matplotlib.pyplot as plt 
import math

file = open('carprices.txt')
next(file)

initial = []
lines = file.readlines()

for val in lines:
    try:
        initial.append(int(val))
        
    except ValueError:
        print('This data '+val+' is invalid and will be ignored')
        lines.remove(val)
        
final = []
for val in initial:
    final.append(math.floor(val/10)) 
    
range = (0, 50)
bins = 50

# plotting a histogram 
plt.hist(final, bins, range,color = 'cornflowerblue', histtype = 'bar', rwidth = 0.8) 

# x-axis label 
plt.xlabel('Price in tens of thousands') 

# frequency label 
plt.ylabel('Number of cars') 

# plot title 
plt.title('Frequency of Car Prices') 

# function to show the plot 
plt.show()    
```

```python
import matplotlib.pyplot as plt
import numpy as np

file=open("cartype.txt","r")
types=['sporty','van','compact','large','midsize','small']
frequency=[0]*6
data=file.read().split('\n')
file.close()

for car_type in data:
    if car_type in types:
        i=types.index(car_type)
        frequency[i]+=1
        
colors=['skyblue','blue','purple','lightskyblue','blue','purple']
plt.pie(frequency, labels=types, colors=colors,autopct='%1.2f%%',
        explode = (0.025,0.025,0.025,0.025,0.025,0.025),shadow=False, startangle=80)
plt.axis('auto')
plt.show()
```
