```python
#define your own Vehicle class here
#the order of the four required positional arguments should be: make,model,year,mileage
#please use the conventional way to define function names and refer to the following test to adjust function names 
#no need to handle exceptions here

class Vehicle:
    
    def __init__(self,make,model,year,mileage):
        self.__make=make
        self.__model=model
        self.__year=year
        self.__mileage=mileage
        
    def get_make(self):
        return self.__make
    
    def set_make(self,make):
        self.__make=make
        
    def get_model(self):
        return self.__model
    
    def set_model(self,model):
        self.__model=model
        
    def get_year(self):
        return self.__year
    
    def set_year(self,year):
        self.__year=year
    
    def get_mileage(self):
        return self.__mileage
    
    def set_mileage(self):
        self.__mileage=mileage
        
    def __str__(self):
        return "Make: " + self.__make + "; Model: " + self.__model + "; Year of Manufacture: " + str(self.__year) + "; Mileage: " + str(self.__mileage)
    
```

```python
#define your own Employee class here
#the order of the three required positional arguments should be: name,address,vehicle object
#please use the conventional way to define function names and refer to the following test to adjust function names
#use emp for short of employee in naming attributes and functions
#no need to handle exceptions here

class Employee:
    
    def __init__(self,name,address,vehicle):
        self.__emp_name=name
        self.__emp_addr=address
        self.__vehicle=vehicle
        
    def get_emp_name(self):
        return self.__emp_name
    
    def get_emp_addr(self):
        return self.__emp_addr
      
    def get_vehicle(self):
        return self.__vehicle
    
    def set_emp_name(self,name):
        self.__emp_name=name
    
    def set_emp_addr(self,address):
        self.__emp_addr=address

    def set_vehicle(self,veh):
        self.__vehicle=vehicle
    
    def __str__(self):
        allData = "\nEmployee Name: " + self.__emp_name
        allData += "; Employee Address: " + str(self.__emp_addr)
        allData +=  "\n" + self.__vehicle.__str__()
        return allData
    
        
```

```python
#define your own Full Time Employee class here
#Full Time Employee class inherits from Employee class
#the order of the four required positional arguments should be: name,address,vehicle object,salary
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class FullTimeEmployee(Employee):
    
    def __init__(self,name,address,vehicle,salary):
        Employee.__init__(self,name,address,vehicle)
        self.__salary=salary
        
    def get_salary(self):
        return self.__salary
    
    def set_salary(self,salary):
        self.__salary=salary
    
    def compute_compensation(self):
        if self.__salary<=45000:
            return (self.__salary-(self.__salary*18/100))/52
        elif 45000 < self.__salary <= 82000:
            return (self.__salary-(45000*18/100 + (self.__salary-45000)*28/100))/52
        else:
            return (self.__salary-(45000*18/100 + (82000-45000)*28/100 + (self.__salary-82000)*33/100))/52

    def compute_reimbursement(self,exp):
        
        if exp<=10000:
            reimbursement = exp/52
            return reimbursement
        
        else:
            reimbursement = 10000 + (exp-10000)*0.5
            reimbursement = reimbursement/52
            return reimbursement
        
    def __str__(self):

        childData = super().__str__()
        childData = "\nDetails of this Full Time Employee are:" + childData
        childData += "\nSalary: " + '{0:0.2f}'.format(self.__salary)
        return childData
            
```

```python
#define your own Hourly Employee class here
#Hourly Employee class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,hourly rate
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class HourlyEmployee(Employee):
    
    def __init__(self,name,address,vehicle,hours_worked,hourly_rate):
        Employee.__init__(self,name,address,vehicle)
        self.__hours_wrkd=hours_worked
        self.__hourly_rate=hourly_rate
    
    def get_hours_worked(self):
        return self.__hours_wrkd
    
    def set_hours_worked(self,hours_worked):
        self.__hours_wrkd=hours_worked
    
    def get_hourly_rate(self):
        return self.__hourly_rate
    
    def set_hourly_rate(self,hourly_rate):
        self.__hourly_rate=hourly_rate
        
    def compute_compensation(self):
        if self.__hours_wrkd<=40:
            return self.__hours_wrkd*self.__hourly_rate*1.0
        else:
            return 40*self.__hourly_rate + (self.__hours_wrkd-40)*self.__hourly_rate*1.8
    
    def compute_reimbursement(self,exp):
        if exp<=100:
            return exp
        else:
            return 100
        
    def __str__(self):
        childData = super().__str__()
        childData = "\nDetails of this Hourly Employee are:" + childData
        childData += "\nHours Worked: " + str(self.__hours_wrkd)
        childData += "; Hourly Rate: " + str(self.__hourly_rate)
        return childData
```

```python
#define your own Consultant class here
#Consultant class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class Consultant(Employee):
    def __init__(self,name,address,vehicle,hours_worked,projecttype):
        Employee.__init__(self,name,address,vehicle)
        self.__hours_wrkd=hours_worked
        self.__project_type=projecttype
    
    def get_hours_worked(self):
        return self.__hours_wrkd
    
    def set_hours_worked(self,hours_worked):
        self.__hours_wrkd=hours_worked
    
    def get_project_type(self):
        return self.__project_type
    
    def set_project_type(self,projecttype):
        self.__project_type=projecttype
    
    def compute_compensation(self):
        if self.__project_type==1:
            return 55*self.__hours_wrkd
        elif self.__project_type==2:
            return 70*self.__hours_wrkd
        else:
            return 85*self.__hours_wrkd
    
    def compute_reimbursement(self,exp):
        if self.__project_type==1:
            return exp
        elif self.__project_type==2:
            return 0.9*exp
        else:
            return 0.8*exp
    
    def __str__(self):
        childData = super().__str__()
        childData = "\nDetails of this Consultant are:" + childData
        childData += "\nHours Worked: " + str(self.__hours_wrkd)
        childData += "; Project Type: " + str(self.__project_type)
        return childData
```

```python
#define your own Management class here
#Management class inherits from both Full Time Employee class and Consultant class
#the order of the six required positional arguments should be: name, address,vehicle object, salary, hours worked, project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class Management(FullTimeEmployee,Consultant):
    def __init__(self,name,address,vehicle,salary,hours_worked,projecttype):
        FullTimeEmployee.__init__(self,name,address,vehicle,salary)
        Consultant.__init__(self,name,address,vehicle,hours_worked,projecttype)
        self.__projecttype=projecttype
    
    #self.__salary=FullTimeEmployee.get_salary(self)
    #self.__hours_worked=Consultant.get_hours_worked(self)
    
        
    def get_salary(self):
        return FullTimeEmployee.get_salary(self)
    
    def get_hours_worked(self):
        return Consultant.get_hours_worked(self)
    
    def get_project_type(self):
        return self.__projecttype
    
    def compute_compensation(self):
        if self.__projecttype==1:
            Consultant.set_project_type(self, 1)
            c=Consultant.compute_compensation(self)
            f=FullTimeEmployee.compute_compensation(self)
            return c+f
        elif self.__projecttype==2:
            Consultant.set_project_type(self, 2)
            c=Consultant.compute_compensation(self)
            f=FullTimeEmployee.compute_compensation(self)
            return c+f
        elif self.__projecttype==3:
            Consultant.set_project_type(self, 3)
            c=Consultant.compute_compensation(self)
            f=FullTimeEmployee.compute_compensation(self)
            return c+f
        else:
            return 0
    
    def compute_reimbursement(self,fullexpense,consultexpense):
        self.__fullexpense=fullexpense
        self.__consultexpense=consultexpense
        c=(FullTimeEmployee.compute_reimbursement(self,self.__fullexpense)+Consultant.compute_reimbursement(self,self.__consultexpense))
        return c
    
    def __str__(self):
        k='Details of this Management are:'
        k+='\n'
        k+=Employee.__str__(self)
        k+='\n'
        k+='Salary: {0:.2f};Hours Worked: {1};Project Type: {2}'.format(self.get_salary(),self.get_hours_worked(),self.__projecttype)
        return k
```

```python
#complete this get_vehicle_input() function to prepare for the main menu application
#this function is used to ask for vehicle information and then return them: make,model,year,mileage
#this function doesn't take arguement(s)
#no need to handle exceptions of make input and model input
#handle exceptions to make sure year is a 4-digit positive number between 1900 and 2020
#handle exceptions to make sure mileage is a positive number
class YearException(Exception):
    
    def __str__(self):
        return "Please enter an integer value for year in the format of yyyy between 1900 and 2020."
    
class MileageException(Exception):
    
    def __str__(self):
        return "Mileage cannot be negative"

def get_vehicle_input():
    vehicle_make = input('Enter the vehicle make: ')
    vehicle_model = input('Enter the vehicle model: ')

    while(1):
        try:
            vehicle_year=int(input('Enter the year of manufacture(yyyy):'))
            while(1900<=vehicle_year<=2020):
                try:
                    vehicle_mileage=int(input('Enter the mileage:'))
                    if vehicle_mileage>=0:
                        return (vehicle_make,vehicle_model,vehicle_year,vehicle_mileage)
                        break
                    else:
                        raise MileageException
                
                except MileageException:
                    print('Mileage cannot be negative')
            raise YearException
        except YearException:
            print('Please enter an integer value for year in the format of yyyy between 1900 and 2020.')
```

```python
#complete this read_file_data() function to prepare for the main menu application
#this function is used to read information from an existing databse/binary file, which is empdata.dat in the final exam
#this function doesn't take argument(s)
#this function returns a list containing all employee objects stored in the database; each employ object is an element in the list; the list will also be used and updated later
#need to handle exceptions if file doesn't exist
import pickle
def read_file_data():
    employee=[]
    with open('empdata.dat','rb') as data:
        try:
            while(True):
                employee.append(pickle.load(data))
        except EOFError:
            pass
    return employee
```

```python
#complete option 5 related code here
#the function is used to show some basic statistics. They are: 
#(1) the number of employees stored in the database
#(2) the highest weekly compensation
#(3) the mean weekly compensation
#(4) the number of employees who have a vehicle with over 100,000 mileage
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function doesn't return anything
#no need to handle exceptions
import statistics as stats
def run_option5(my_lst):
    #type in your code
        print("\n============================================================")        
        print("Below is the statistics of all employees in this database.")
        print("==============================================================") 
        
        print('There are {0} employees stored in this employee database'.format(len(my_lst)))
        compensation = []
        
        for val in my_lst:
            compensation.append(val.compute_compensation())
            
        compensation.sort(reverse=True)
        mileage=0
        for i in range(len(my_lst)):
            if my_lst[i].get_vehicle().get_mileage()>100000:
                mileage+=1
        print('The highest weekly compensation is ${0}'.format(compensation[0]))
        print('The mean weekly compensation is: ${0}'.format(stats.mean(compensation)))
        print('The number of employees who have a vehicle with over 100,000 mileage is {0}.'.format(mileage))
    
```
