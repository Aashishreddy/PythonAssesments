```python
class Address:
    def __init__(self,street,city,state,zipcode):
        self.__street=street
        self.__city=city
        self.__state=state
        self.__zipcode=zipcode
    
    def get_street(self):
        return self.__street
    def get_city(self):
        return self.__city
    def get_state(self):
        return self.__state
    def get_zipcode(self):
        return self.__zipcode
    def set_street(self,street):
        self.__street=street 
    def set_city(self,city):
        self.__city=city    
    def set_state(self,state):
        self.__state=state
    def set_zipcode(self,zipcode):
        self.__zipcode=zipcode
    
    def __str__(self):
        str="Employee Address: {0},{1},{2},{3}".format(self.__street,self.__city,self.__state,self.__zipcode)
        str+="\n"
        return str
    
class Vehicle:
    def __init__(self,make,model,year,mileage):
        self.__make=make
        self.__model=model
        self.__year=year
        self.__mileage=mileage
        
    def get_make(self):
        return self.__make
    def get_model(self):
        return self.__model
    def get_year(self):
        return self.__year
    def get_mileage(self):
        return self.__mileage
    def set_make(self,make):
        self.__make=make
    def set_model(self,model):
        self.__model=model
    def set_year(self,year):
        self.__year=year    
    def set_mileage(self):
        self.__mileage=mileage
        
    def __str__(self):
        str="Make: {0}; Model: {1};Year of Manufacture: {2};Mileage: {3}".format(self.__make,self.__model,self.__year,self.__mileage)
        return str
    
class Employee:
    
    def __init__(self,name,address,vehicle):
        self.__name=name
        
        address = address.split(",")
        self.__address = Address(address[0],address[1],address[2],address[3])
        
        vehicle = vehicle.split(",")
        self.__vehicle = Vehicle(vehicle[0],vehicle[1],vehicle[2],vehicle[3])
        
    def get_name(self):
        return self.__name
    def get_year(self):
        return self.__vehicle.get_year()
    def set_name(self,name):
        self.__name=name
    def mileage(self):
        return self.__vehicle.get_mileage()
    
    def __str__(self):
        str = "Employee Details are:"
        str+="\n"
        str+="Employee Name: {0}".format(self.__name)
        str+=";"
        str+=self.__address.__str__()
        str+=self.__vehicle.__str__()
        return str

class FullTimeEmployee(Employee):
    def __init__(self,name,jobtype,address,vehicle,salary):
        super().__init__(name,address,vehicle)
        self.__jobtype=jobtype
        self.__salary=salary
        
    def get_jobtype(self):
        return self.__jobtype
    def get_name(self):
        return self.__salary
    def set_jobtype(self,jobtype):
        self.__jobtype=jobtype
    
    def set_name(self,salary):
        self.__salary=salary
    def car_mileage(self):
        return int(super().mileage())
    def new_car(self):
        return str(super().get_name())+","+str(super().get_year())
    
    def weekly_compensation(self):
        if self.__salary<=45000:
            return str(super().get_name())+','+str((self.__salary-(self.__salary*18/100))/52)
        elif 45000<self.__salary<=82000:
            return str(super().get_name())+','+str((self.__salary-(45000*18/100 + (82000-45000)*28/100))/52)
        else:
            return str(super().get_name())+','+str((self.__salary-(45000*18/100 + (82000-45000)*28/100 + (self.__salary-82000)*33/100))/52)
    
#     def get_compensation(salary):
#         if salary < 45000:
#             taxes = salary * 18/100
#         if salary > 45000 and salary < 82000:
#             taxes = (45000 * 18/100) + (salary - 45000) * 28/100 
#         if salary > 82000:
#             taxes = (45000 * 18/100) + (37000 * 28/100) + (salary - 82000) * 33/100
            
#         weekly_compensation = (salary - taxes)/52
#         return weekly_compensation    
    def __str__(self):
        str=super().__str__()
        str+="\n"
        str+="Salary: {0}".format(self.__salary)
        str+="\n"
        return str

class HourlyEmployee(Employee):
    def __init__(self,name,jobtype,address,vehicle,hours,rate):
        super().__init__(name,address,vehicle)
        self.__jobtype=jobtype
        self.__hours=hours
        self.__rate=rate

    def get_jobtype(self):
        return self.__jobtype
    def get_hours(self):
        return self.__hours
    def get_rate(self):
        return self.__rate
    
    def set_jobtype(self,jobtype):
        self.__jobtype=jobtype
    def set_hours(self,hours):
        self.__hours=hours
    def set_rate(self,rate):
        self.__rate=rate
    def car_mileage(self):
        return int(super().mileage())
    def new_car(self):
        return str(super().get_name())+","+str(super().get_year())
    
    def weekly_compensation(self):
        if self.__hours<=40:
            return str(super().get_name())+','+str(self.__hours*self.__rate)
        else:
            return str(super().get_name())+','+str(40*self.__rate + (self.__hours-40)*self.__rate*1.8)
     
#       def get_compensation(hoursWorked, hourlyRate):
#         if hoursWorked <= 40:
#             weekly_compensation = hoursWorked * hourlyRate
#         if hoursWorked > 40:
#             weekly_compensation = (40 * hourlyRate) + ((hoursWorked - 40) * hourlyRate * 1.8)
            
#         return weekly_compensation   
    
    def __str__(self):
        str=super().__str__()
        str+="\n"
        str+="Hours Worked: {0}; Hourly Rate: {1}".format(self.__hours,self.__rate)
        str+="\n"
        return str

class Consultant(Employee):
    def __init__(self,name,jobtype,address,vehicle,hours,projecttype):
        super().__init__(name,address,vehicle)
        self.__jobtype=jobtype
        self.__hours=hours
        self.__projecttype=projecttype

    def get_jobtype(self):
        return self.__jobtype
    
    def get_hours(self):
        return self.__hours    
   
    def car_mileage(self):
        return int(super().mileage())
    
    def set_jobtype(self,jobtype):
        self.__jobtype=jobtype
    
    def set_hours(self,hours):
        self.__hours=hours
    
    def get_rate(self):
        return self.__projecttype
    
    def set_rate(self,projecttype):
        self.__projecttype=projecttype
     
    def new_car(self):
        return str(super().get_name())+","+str(super().get_year())
    
    def weekly_compensation(self):
        if self.__projecttype==1:
            return str(super().get_name())+','+str(55*self.__hours)
        elif self.__projecttype==2:
            return str(super().get_name())+','+str(70*self.__hours)
        else:
            return str(super().get_name())+','+str(85*self.__hours)
        
#         def get_compensation(hoursWorked, ProjectType):
#         if ProjectType == 1:
#             hourlyrate = 55
            
#         if ProjectType == 2:
#             hourlyRate = 70
            
#         if ProjectType == 3:
#             hourlyRate = 85
        
#         weekly_compensation = hoursWorked * hourlyrate
#         return weekly_compensation    
    
    def __str__(self):
        str=super().__str__()
        str+="Hours Worked: {0}; Project Type: {1}".format(self.__hours,self.__projecttype)
        str+="\n"
        return str
    
    
def employee_data():
    fulltime1=FullTimeEmployee("Amy"," Full"," 100 W Campbell Road,Richardson,Texas,75080"," Honda,Civic,2014,50000", 60000)
    fulltime2=FullTimeEmployee("Bob"," Full"," 200 W Campbell Road,Richardson,Texas,75080"," Toyota,Camry,2010,60000", 80000)
    fulltime3=FullTimeEmployee("Evan"," Full"," 300 W Campbell Road,Richardson,Texas,75080"," BMW,X3,2016,30000", 120000)
    hourly4=HourlyEmployee("Grace"," Hourly"," 400 W Campbell Road,Richardson,Texas,75080"," Kia,Rio,2005,150000",50,20)
    hourly5=HourlyEmployee("James"," Hourly"," 500 W Campbell Road,Richardson,Texas,75080"," Mazda,Mazda3,2010,80000",35,25)
    hourly6=HourlyEmployee("Luna"," Hourly"," 600 W Campbell Road,Richardson,Texas,75080"," Nissan,Sentra,2005,160000",20,15)
    con7=Consultant("Michael"," Consultant"," 700 W Campbell Road,Richardson,Texas,75080"," Ford,Ranger,2011,80000",40,1)
    con8=Consultant("Tom"," Consultant"," 800 W Campbell Road,Richardson,Texas,75080"," Mercedes-Benz,GLE350,2018,30000",50,2)
    con9=Consultant("Zoey"," Consultant"," 900 W Campbell Road,Richardson,Texas,75080"," Tesla,Model3,2018,15000",30,3)
    
    data = [fulltime1,fulltime2,fulltime3,hourly4,hourly5,hourly6,con7,con8,con9]
    return data


def main():
    
    employee = employee_data()
    for i in employee:
        print(i)

main()    
```

```python
import statistics

def highest_salary():
    compensation = []
    compensationdict = {}
    emp=employee_data()
    for i in emp:
        compensation.append(i.weekly_compensation())
        
    for i in compensation:
        splii=i.split(",")
        compensationdict.update({splii[0]:float(splii[1])})
        
    dictvalues=list(compensationdict.values())
    dictvalues.sort(reverse=True)
    
    final=[]
    for i,j in compensationdict.items():
        if j==dictvalues[0]:
            final.append(i)
    print("The people who have the highest weekly compensation ${0} are: {1}".format(int(dictvalues[0]),final))

highest_salary()

def comp():
    comp = []
    compdict = {}
    emp=employee_data()
    for i in emp:
        comp.append(i.weekly_compensation())
        
    for i in comp:
        s1=i.split(",")
        compdict.update({s1[0]:float(s1[1])})
    
    dictvalues=list(compdict.values())
    print("The Mean weekly compensation of all objects is: {:0.2f}".format(sum(dictvalues)/len(dictvalues)))

comp()

def newvehicle():
    vehicle=[]
    vehicledict={}
    emp=employee_data()
    for i in emp:
        vehicle.append(i.new_car())
    for i in vehicle:
        k=i.split(",")
        vehicledict.update({k[0]:int(k[1])})
    
    vehiclevalues=list(vehicledict.values())
    vehiclevalues.sort(reverse=True)
    
    final=[]
    for i,j in vehicledict.items():
        if j==vehiclevalues[0]:
            final.append(i)
    print("The people who have the newest vehicle are: {0}".format(final))
        

newvehicle()

def mileage():
    mileage = []
    employee=employee_data()
    for i in employee:
        mileage.append(i.car_mileage())
    print("The Mean mileage of all Vehicles is: {:0.2f}".format(statistics.mean(mileage)))

mileage()
```
